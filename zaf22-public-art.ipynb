The dataset I used to describe the best neighborhood in Pittsburgh is the one with the most public art projects.

```python
#imports
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Data is being read through csv file
parks = pd.read_csv("https://data.wprdc.org/datastore/dump/00d74e83-8a23-486e-841b-286e1332a151",
                           index_col="title", 
                           parse_dates=True)
#Lists the amount of times each neighborhood had a public art project in the dataset.
parks['neighborhood'].value_counts()
```

As you can see, the Central Business District has the most public art in Pittsburgh with 27 pieces. Quite a few cities share the lowest amount of 1, including Sheraden and Stanton Heights.

```python
#A bar graph of the amount of art projects in each neighborhood.
parks['neighborhood'].value_counts().plot(kind='bar', width=.4, figsize=(25,10), position=0)
```

Above is a bar graph of the amount of art projects in each neighborhood ranked from highest to lowest. I figured this would be easier to look at and understand.
